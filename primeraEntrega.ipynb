con “buenos días”, a partir del medio día saludaremos con “Buenas tardes”, y a partir de las 21 y hasta las 5 de la mañana siguiente
saludaremos con buenas noches.


```python
def saludos_email(h):
  saludo = None
  if h >= 5 and h < 12:
    saludo = "Buenos dias"
  elif h >= 12 and h < 21:
    saludo = "Buenas tardes"
  else:
    saludo = "Buenas noches"
  return saludo

assert saludos_email(6) == "Buenos dias", "Error {}".format(saludos_email(6))
assert saludos_email(12) == "Buenas tardes", "Error {}".format(saludos_email(12))
assert saludos_email(21) == "Buenas noches", "Error {}".format(saludos_email(21))

print("Introduce la hora")
hora = int(input())
print(saludos_email(hora))
```

(sin tilde).

```python
def triangulo(a, b, c):
  tipo_triangulo = None
  if a == b and b == c:
    tipo_triangulo =  "equilatero"
  elif a != b and a != c and b != c:
    tipo_triangulo =  "escaleno"
  else:
    tipo_triangulo =  "isosceles"
  return tipo_triangulo

assert triangulo(1, 2, 3) == "escaleno", "Error {}".format(triangulo(1, 2, 3))
assert triangulo(2, 2, 2) == "equilatero", "Error {}".format(triangulo(2, 2, 2))
assert triangulo(2, 2, 1) == "isosceles", "Error {}".format(triangulo(2, 2, 1))


print(triangulo(1, 3, 2))
print(triangulo(1, 1, 2))
print(triangulo(1, 1, 1))
```

la función "cercanias(linea, sentido)" que reciba como parámetros un número entero positivo que represente la línea de cercanías y un

Para determinar si un año es bisiesto, es necesario que cumpla las siguientes condiciones: el año debe ser divisible entre 4 pero no entre
100, o bien, debe ser divisible entre 400. Por ejemplo, el año 2024 es bisiesto debido a que cumple la condición de ser divisible entre 4.
Por contra, 2023 no es bisiesto ya que no es divisile por 4 ni por 400.
Crea la función es_bisiesto(a), que dado un año a devuelve True en caso de que sea bisiesto y False en caso de que no lo sea

```python
def es_bisiesto(a):
  anyo = None
  if (a % 4 == 0 and a % 100 != 0) or (a % 400 == 0):
    anyo = True
  else:
    anyo = False
  return anyo
assert es_bisiesto(2023) == False, "Error {}".format(es_bisiesto(2023))
assert es_bisiesto(2024) == True, "Error {}".format(es_bisiesto(2024))
assert es_bisiesto(1900) == False, "Error {}".format(es_bisiesto(1900))


print(es_bisiesto(2000))
print(es_bisiesto(2001))
print(es_bisiesto(2002))
print(es_bisiesto(2003))
print(es_bisiesto(2004))
```

mensual que una vez superado ya no debes pagar.
Crea una función farmacia(usuario, salario_bruto) que dados estos dos valores devuelva el porcentaje que debes pagar del
precio de los medicamentos y el máximo mensual, esto se codifica con -1.
Los trabajadores si tienen un salario inferior a los 20000€ deben pagar un 45%. Si cobran entre 20000 y 49999 deben pagar un 65%, y si
cobran más deben pagar un 70%. Los trabajadores no tienen un límite mensual.

```python
def farmacia(usuario, salario_bruto):
  descuento = None
  maximo = None
  if usuario == "trabajador":
    if salario_bruto < 20000:
      descuento = 45
      maximo = -1
      return descuento, maximo
    elif salario_bruto >= 20000 and  salario_bruto < 49999: # elif 20000 <= salario_bruto < 49999:
      descuento = 65
      maximo = -1
      return descuento, maximo
    else:
      descuento = 70
      maximo = -1
      return descuento, maximo
  elif usuario == "pensionista":
    if salario_bruto < 20000:
      descuento = 10
      maximo = 7
      return descuento, maximo
    elif salario_bruto >= 20000 and  salario_bruto < 40000: # elif 20000 <= salario_bruto < 40000:
      descuento = 10
      maximo = 25
      return descuento, maximo
    else:
      descuento = 50
      maximo = 100
      return descuento, maximo
  else:
      descuento = 0
      maximo = 0
      return descuento, maximo

print("Pensionista imprimir (10, 7)", farmacia("pensionista", 12000))
print("Trabajador imprimir (45, -1 )", farmacia("trabajador", 12000))
print("Accidente imprimir (0, 0)", farmacia("accidente", 12000))
print("Discapacidad imprimir (0, 0 )", farmacia("discapacidad", 50000))
print("Accidente imprimir (0, 0 )", farmacia("accidente", 50000))

print("Trabajados imprimir (45, -1 )", farmacia("trabajador", 15000))
print("Trabajados imprimir (65, -1 )", farmacia("trabajador", 25000))
print("Trabajados imprimir (70, -1 )", farmacia("trabajador", 60000))

print("Pensionista imprimir (10, 7)", farmacia("pensionista", 15000))
print("Pensionista imprimir (10, 25)", farmacia("pensionista", 25000))
print("Pensionista imprimir (50, 100)", farmacia("pensionista", 45000))

print("Accidente imprimir (0, 0)", farmacia("accidente", 15000))
print("Accidente imprimir (0, 0)", farmacia("accidente", 25000))
print("Accidente imprimir (0, 0)", farmacia("accidente", 45000))
```
